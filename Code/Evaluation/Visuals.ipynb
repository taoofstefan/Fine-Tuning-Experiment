```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Baseline data
baseline_data = {
    'Model Name': ['Llama 2', 'Mistral Instruct', 'gpt-3.5', 'gpt-4-turbo', 'gpt-4o'],
    'Accuracy': [21, 21, 87, 91, 94],
    'Precision': [0, 0, 5, 26, 31],
    'TCE': [4, 21, 7, 15, 15]
}

# Create DataFrame
baseline_df = pd.DataFrame(baseline_data)

# Plotting Accuracy
plt.figure(figsize=(10, 5))
bars = plt.bar(baseline_df['Model Name'], baseline_df['Accuracy'], color='skyblue')
plt.title('Baseline Evaluation: Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')
plt.show()

# Plotting Precision
plt.figure(figsize=(10, 5))
bars = plt.bar(baseline_df['Model Name'], baseline_df['Precision'], color='lightgreen')
plt.title('Baseline Evaluation: Precision')
plt.ylabel('Precision')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')
plt.show()

# Plotting TCE
plt.figure(figsize=(10, 5))
bars = plt.bar(baseline_df['Model Name'], baseline_df['TCE'], color='salmon')
plt.title('Baseline Evaluation: TCE')
plt.ylabel('TCE')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')
plt.show()
```

```python
# Data preparation
models = ['Llama 2', 'Llama 2 FT', 'Mistral Instruct', 'Mistral Instruct FT', 'gpt-3.5', 'gpt-4-turbo', 'gpt-4o']
accuracy_values = [21, 77, 21, 83, 87, 91, 94]

# Create DataFrame
accuracy_df = pd.DataFrame({
    'Model': models,
    'Accuracy': accuracy_values
})

# Plotting Accuracy
plt.figure(figsize=(10, 5))
bars = plt.bar(accuracy_df['Model'], accuracy_df['Accuracy'], color=['skyblue', 'deepskyblue', 'lightgreen', 'limegreen', 'grey', 'grey', 'grey'])
plt.title('Accuracy Comparison Before and After Fine-Tuning')
plt.ylabel('Accuracy')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')
plt.xticks(rotation=45)
plt.show()

# Data preparation
precision_values = [0, 35, 0, 50, 5, 26, 31]

# Create DataFrame
precision_df = pd.DataFrame({
    'Model': models,
    'Precision': precision_values
})

# Plotting Precision
plt.figure(figsize=(10, 5))
bars = plt.bar(precision_df['Model'], precision_df['Precision'], color=['skyblue', 'deepskyblue', 'lightgreen', 'limegreen', 'grey', 'grey', 'grey'])
plt.title('Precision Comparison Before and After Fine-Tuning')
plt.ylabel('Precision')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')
plt.xticks(rotation=45)
plt.show()

# Data preparation
tce_values = [4, 9, 21, 4, 7, 15, 15]

# Create DataFrame
tce_df = pd.DataFrame({
    'Model': models,
    'TCE': tce_values
})

# Plotting TCE
plt.figure(figsize=(10, 5))
bars = plt.bar(tce_df['Model'], tce_df['TCE'], color=['skyblue', 'deepskyblue', 'lightgreen', 'limegreen', 'grey', 'grey', 'grey'])
plt.title('TCE Comparison Before and After Fine-Tuning')
plt.ylabel('TCE')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')
plt.xticks(rotation=45)
plt.show()
```

```python
# Data preparation
models = ['Llama 2', 'Llama 2 FT', 'Mistral Instruct', 'Mistral Instruct FT', 'gpt-3.5-turbo', 'gpt-4-turbo', 'gpt-4o']
accuracy_values = [21, 77, 21, 83, 87, 91, 94]
precision_values = [0, 35, 0, 50, 5, 26, 31]
tce_values = [4, 9, 21, 4, 7, 15, 15]

# Maximum values
max_accuracy = 100
max_precision = 79
max_tce = 21

# Plotting Accuracy with Maximum Reference Line
plt.figure(figsize=(12, 6))
bars = plt.bar(models, accuracy_values, color=['skyblue', 'deepskyblue', 'lightgreen', 'limegreen', 'grey', 'grey', 'grey'])
plt.axhline(max_accuracy, color='red', linestyle='dashed', linewidth=2, label='Max Accuracy (100)')
plt.title('Accuracy Comparison with Maximum Reference Line')
plt.ylabel('Accuracy')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')
plt.legend()
plt.xticks(rotation=45)
plt.plot()

# Plotting Precision with Maximum Reference Line
plt.figure(figsize=(12, 6))
bars = plt.bar(models, precision_values, color=['skyblue', 'deepskyblue', 'lightgreen', 'limegreen', 'grey', 'grey', 'grey'])
plt.axhline(max_precision, color='red', linestyle='dashed', linewidth=2, label='Max Precision (79)')
plt.title('Precision Comparison with Maximum Reference Line')
plt.ylabel('Precision')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')
plt.legend()
plt.xticks(rotation=45)
plt.plot()

# Plotting TCE with Maximum Reference Line
plt.figure(figsize=(12, 6))
bars = plt.bar(models, tce_values, color=['skyblue', 'deepskyblue', 'lightgreen', 'limegreen', 'grey', 'grey', 'grey'])
plt.axhline(max_tce, color='red', linestyle='dashed', linewidth=2, label='Max TCE (21)')
plt.title('TCE Comparison with Maximum Reference Line')
plt.ylabel('TCE')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')
plt.legend()
plt.xticks(rotation=45)
plt.plot()
```

```python
# Calculate percentage improvement
percentage_improvement = {
    'Model': ['Llama 2', 'Mistral Instruct'],
    'Accuracy Improvement': [(77-21)/21*100, (83-21)/21*100],
    'Precision Improvement': [35/1*100, 50/1*100],
    'TCE Improvement': [(9-4)/4*100, (4-21)/21*100]
}

# Create DataFrame
improvement_df = pd.DataFrame(percentage_improvement)

# Plotting Percentage Improvement for Accuracy
plt.figure(figsize=(8, 5))
bars = plt.bar(improvement_df['Model'], improvement_df['Accuracy Improvement'], color=['skyblue', 'lightgreen'])
plt.title('Percentage Improvement in Accuracy After Fine-Tuning')
plt.ylabel('Percentage Improvement')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.1f}%', ha='center', va='bottom')
plt.show()

# Plotting Percentage Improvement for Precision
plt.figure(figsize=(8, 5))
bars = plt.bar(improvement_df['Model'], improvement_df['Precision Improvement'], color=['skyblue', 'lightgreen'])
plt.title('Percentage Improvement in Precision After Fine-Tuning')
plt.ylabel('Percentage Improvement')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.1f}%', ha='center', va='bottom')
plt.show()

# Plotting Percentage Improvement for TCE
plt.figure(figsize=(8, 5))
bars = plt.bar(improvement_df['Model'], improvement_df['TCE Improvement'], color=['skyblue', 'lightgreen'])
plt.title('Percentage Improvement in TCE After Fine-Tuning')
plt.ylabel('Percentage Improvement')
plt.xlabel('Model Name')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.1f}%', ha='center', va='bottom')
plt.show()
```
